# Filter rows of a pandas DataFrame by column value

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
movies.shape
```

```python
# create a list in which each element refers to a DataFrame row: True if the row satisfies the condition, False otherwise
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
len(booleans)
```

```python
booleans[0:5]
```

```python
type(booleans)
```

```python
# convert the list to a Series
is_long = pd.Series(booleans)
is_long.head()
```

```python
# use bracket notation with the boolean Series to tell the DataFrame which rows to display
movies[is_long]
```

### in one go

```python
is_long = movies.duration >= 200
movies[is_long]
```

```python
# or simply
movies[movies.duration >= 200]
```

```python
movies.duration >= 200
```

```python
# select the 'genre' Series from the filtered DataFrame
movies[movies.duration >= 200].genre
```

```python
movies[movies.duration >= 200]['genre']
```

```python
# or equivalently, use the 'loc' method. It is used to pick from rows
movies.loc[movies.duration >= 200]
```

```python
# or equivalently, use the 'loc' method. It is used to pick from rows
movies.loc[movies.duration >= 200, 'genre']
```

#  multiple filter

```python
# Let's say we want to see movies that are 200 or longer than 200 minutes and falls in drama genre
movies['genre'].value_counts()
```

```python
# Visualization 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%pylab inline
```

```python
fig = plt.figure(figsize=(20,1))
sns.countplot(y='genre', data=movies);
print(movies.genre.value_counts())
```

```python
movies[(movies.duration >=200) & (movies.genre == 'Drama')]
```

```python
# With or we won't get same results
movies[(movies.duration >=200) | (movies.genre == 'Drama')].head()
```

**Suppose we want to check either genre is crime or drama or action**

```python
# use the '|' operator to specify that a row can match any of the three criteria
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Action')].head(10)
```

```python
# or equivalently, use the 'isin' method
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])].head(10)
```

```python
# Combining all filtering. Movies that are of >=200 minutes and are of crime, action or drama genre
movies[(movies.duration >=200) & (movies.genre.isin(['Crime', 'Drama', 'Action']))]
```
